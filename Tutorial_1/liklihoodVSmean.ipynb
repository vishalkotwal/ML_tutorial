```
def plotting():
  mean = np.arange(5,15,1)
  x = np.arange(0,10,1)
  plt.figure(figsize=(10,10))
  for i in range(len(mean)):
    y = norm.pdf(x,mean[i],1)
    plt.subplot(5,2,i+1)
    plt.scatter(x,y)
    plt.plot(x,y)
    
```

```
plotting()
```

```
def subplotting():
  x_scatter = np.arange(0,11,0.3)
  y_scatter = norm.pdf(x_scatter,5,1)
  list1 = []
  #mean = np.arange(0,11)
  plt.tight_layout()
  plt.scatter(x_scatter,y_scatter,color = 'orangered')
  plt.title('Gaussian Distribution(10 sample points)',fontsize = 15 , fontweight = 'bold')

  mean = np.arange(0,11,1)
  figure,axs = plt.subplots(6,2,figsize = (15,15))
  plt.suptitle("liklihood function for mean (0 to 10)",fontsize= 15,fontweight = "bold")
  
  for i in range(len(mean)):
    y_mean = norm.pdf(x_scatter,mean[i],1)
    result1 = np.prod(y_mean)
    list1.append(result1)
    #print(list1)
    
    if i < 6:
      axs[i,0].set_title('Mean = ' + str(i) +" "+  'varince = 1')
      axs[i,0].scatter(x_scatter,y_scatter,color="orangered")
      axs[i,0].plot(x_scatter,y_mean,color = 'teal')
    else:
      axs[i-6,1].set_title('Mean = ' + str(i) +" " + 'varince = 1')
      axs[i-6,1].scatter(x_scatter,y_scatter,color = "orangered")
      axs[i-6,1].plot(x_scatter,y_mean,color = 'teal')
    plt.tight_layout(pad = 2.5)
  #print(list1)

  ## plotting liklihood vs mean

  plt.title("Liklihood Vs mean")
  plt.plot(mean,list1,color = 'teal')
  plt.scatter(mean,list1,color = 'orangered')
  plt.xlabel('Mean')
  plt.ylabel('Liklihood')
  plt.show()



    









  
```

```
subplotting()
```

```
def liklihood():
  list1 =[]
  x_scatter = np.arange(0,11,0.3)
  mean = np.arange(0,11,1)
  for i in range(len(mean)):
    y_mean = norm.pdf(x_scatter,mean[i],1)
    result1 = np.prod(y_mean)
    list1.append(result1)
  plt.title("Liklihood Vs mean")
  plt.plot(mean,list1,color = 'teal')
  plt.scatter(mean,list1,color = 'orangered')
  plt.xlabel('Mean')
  plt.ylabel('Liklihood')
  plt.show()
  
```
